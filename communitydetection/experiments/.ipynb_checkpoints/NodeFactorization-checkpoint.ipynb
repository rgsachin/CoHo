```python
import numpy as np
from sklearn.decomposition import NMF
```

```python
nodeLookUp={}
nodeIdx=-1
def getNodeId(rawId):
    global nodeIdx
    if rawId in nodeLookUp:
        nodeId = nodeLookUp[rawId]
    else:
        nodeIdx+=1
        nodeId = nodeIdx
        nodeLookUp[rawId] = nodeId
    return nodeId
```

```python
circles = open('C:/Users/sachin/Documents/sanctorium/research/communitydetection/data/circles/1968.circles', 'r')
for circle in circles:
    isFirst= True
    for node in circle.split():
        if(isFirst):
            isFirst=False
            continue
        getNodeId(node)
print nodeLookUp
```

```python
utilMat = np.zeros((len(nodeLookUp),len(nodeLookUp)))
circles = open('C:/Users/sachin/Documents/sanctorium/research/communitydetection/data/circles/1968.circles', 'r')
np.set_printoptions(threshold=1000)
#print utilMat
for circle in circles:
    nodesInCircle = circle.split()
    nodesInCircle = nodesInCircle[1:]
    for node1 in nodesInCircle:
        for node2 in nodesInCircle:
            utilMat[getNodeId(node1),getNodeId(node2)]=100
            utilMat[getNodeId(node2),getNodeId(node1)]=utilMat[getNodeId(node1),getNodeId(node2)]
print utilMat
```

```python
model = NMF(n_components=14,nls_max_iter=10000,solver='cd',l1_ratio=0.3,alpha=0.7,init='nndsvd',tol=0.000001)
#np.set_printoptions(threshold=np.inf,linewidth=500)
print model.fit_transform(utilMat)
#np.set_printoptions(threshold=1000,linewidth=75)
```
